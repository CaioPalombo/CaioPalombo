**P√°ginas seguintes:**

**Anota√ß√£o:** Os t√≠tulos tem a mesma tag, mas os par√°grafos e subt√≠tulos n√£o.

```python
# BIBLIOTECA
import httpx
from bs4 import BeautifulSoup
from re import findall

paginas = ["https://averdade.org.br/category/brasil/",
           "https://averdade.org.br/category/brasil/page/2/",
           ]

def acessar_pagina (link):
  pag_web = httpx.get(link)
  bs = BeautifulSoup(pag_web.text, "html.parser")
  return bs


#loop
for pagina in paginas:
  pagina_inteira = acessar_pagina(pagina)
  lista_noticias = pagina_inteira.find("div", attrs={"id" : "tdi_50"}).find_all("h3", attrs={"class" : "entry-title td-module-title"})
  print(lista_noticias)
  for noticia in lista_noticias:
    titulo = noticia.text.strip()
    print(titulo)
    link_noticia = noticia.a['href']
    print(link_noticia)
```
